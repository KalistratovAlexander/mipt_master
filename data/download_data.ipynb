```python
from pathlib import Path
import gzip
import shutil
import urllib.request

CATEGORY = "Pet_Supplies"
DATA_SPLITTING = "last_out_w_his" # "timestamp_w_his" , "last_out_w_his"
CORE = 5 # 5, 0

BASE = "https://mcauleylab.ucsd.edu/public_datasets/data/amazon_2023"
DATA_DIR = Path.cwd() / "data"
DATA_DIR.mkdir(parents=True, exist_ok=True)

FILES = {
    f"{CATEGORY}.train.csv": f"{BASE}/benchmark/{CORE}core/{DATA_SPLITTING}/{CATEGORY}.train.csv.gz",
    f"{CATEGORY}.valid.csv": f"{BASE}/benchmark/{CORE}core/{DATA_SPLITTING}/{CATEGORY}.valid.csv.gz",
    f"{CATEGORY}.test.csv":  f"{BASE}/benchmark/{CORE}core/{DATA_SPLITTING}/{CATEGORY}.test.csv.gz",
    f"{CATEGORY}.rating_only.csv": f"{BASE}/benchmark/{CORE}core/rating_only/{CATEGORY}.csv.gz",
    f"meta_{CATEGORY}.jsonl":   f"{BASE}/raw/meta_categories/meta_{CATEGORY}.jsonl.gz",
    # f"review_{CATEGORY}.jsonl": f"{BASE}/raw/review_categories/{CATEGORY}.jsonl.gz",
}
```

```python
def download_and_unzip(url: str, out_path: Path) -> Path:
    """Download .gz to disk, unzip to out_path, then remove the .gz file."""
    if out_path.exists():
        print(f"Exists: {out_path}")
        return out_path

    gz_path = out_path.with_suffix(out_path.suffix + ".gz")
    print(f"Downloading: {url}")
    urllib.request.urlretrieve(url, gz_path)

    print(f"Unzipping: {gz_path} -> {out_path}")
    with gzip.open(gz_path, "rb") as f_in, open(out_path, "wb") as f_out:
        shutil.copyfileobj(f_in, f_out)

    gz_path.unlink(missing_ok=True)
    print(f"Done: {out_path}\n")
    return out_path
```

```python
print(f"Workdir: {Path.cwd()}")
print(f"Data dir: {DATA_DIR}")
print(f"Category: {CATEGORY}\n")

paths = {name: download_and_unzip(url, DATA_DIR / name) for name, url in FILES.items()}
paths
```
